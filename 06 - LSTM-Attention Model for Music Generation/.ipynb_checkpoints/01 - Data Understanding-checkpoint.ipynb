### Load convert and play example data

```python
from IPython.display import Audio
from music21 import converter
from midi2audio import FluidSynth

path = "midi_dataset"
midi_filepath = f'{path}/Piano Sonata No.27.mid'
wav_filepath = midi_filepath.replace('.mid', '.wav')
midi_score = converter.parse(midi_filepath).chordify()

# convert midi
FluidSynth().midi_to_audio(midi_filepath, wav_filepath)

Audio(wav_filepath)
```

### Show example data in text form

```python
print(midi_score.show('text'))
```

### Show example data in piano format

```python
print(midi_score.show())
```
