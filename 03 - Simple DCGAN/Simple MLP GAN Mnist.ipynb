```python
import math
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from IPython.display import display_html
from tensorflow.keras.datasets import mnist
```

## Parametres:

```python
latent_size = 100
batch_size = 32
lr = 2e-4
decay = 6e-8
epochs = 10
random_normal_dimensions = (latent_size)
```

## Load Data:

```python
(x_train, _), (_, _) = mnist.load_data()

x_train = x_train.astype(np.float32) / 255
dataset = tf.data.Dataset.from_tensor_slices(x_train).batch(batch_size=batch_size)
```

## Generator:

```python
g = tf.keras.layers.Input(shape=(random_normal_dimensions, ))
x = tf.keras.layers.Dense(64, activation="selu")(g)
x = tf.keras.layers.Dense(128, activation="selu")(x)
x = tf.keras.layers.Dense(28 * 28, activation="sigmoid")(x)
o = tf.keras.layers.Reshape([28, 28])(x)
generator = tf.keras.models.Model(g, o, name="generator")

tf.keras.utils.plot_model(generator, "generator.png", show_shapes=True, dpi=60)
```

## Discriminator:

```python
d = tf.keras.layers.Input(shape=(28, 28))
x = tf.keras.layers.Flatten()(d)
x = tf.keras.layers.Dense(128, activation="selu")(x)
x = tf.keras.layers.Dense(64, activation="selu")(x)
o = tf.keras.layers.Dense(1, activation="sigmoid")(x)
discriminator = tf.keras.models.Model(d, o, name="discriminator")

tf.keras.utils.plot_model(discriminator, "discriminator.png", show_shapes=True, dpi=60)
```

## Adversarial:

```python
adversarial = tf.keras.models.Model(
    g, 
    discriminator(generator(g)),
    name="adversarial"
)

tf.keras.utils.plot_model(adversarial, "adversarial.png", show_shapes=True, dpi=60)
```

```python
discriminator.compile(
    loss="binary_crossentropy", 
    optimizer =tf.keras.optimizers.RMSprop(lr=lr, decay=decay),
    metrics=["accuracy"]                     
)

adversarial.compile(
    loss="binary_crossentropy", 
    optimizer = tf.keras.optimizers.RMSprop(lr=lr * 0.5, decay=decay * 0.5),
    metrics=["accuracy"]
)
```

```python
epochs = 100
score = {"a": [], "d": []}
for epoch in range(epochs):
    for real_images in dataset:
        noise = tf.random.normal(shape=[batch_size, random_normal_dimensions])
        fake_images = generator.predict(noise)
        mixed_images = tf.concat([fake_images, real_images], axis=0)
        discriminator_labels = tf.constant([[0.]] * batch_size + [[1]] * batch_size)
        discriminator.trainable = True
        dloss, dacc = discriminator.train_on_batch(mixed_images, discriminator_labels)

        noise = tf.random.normal(shape=[batch_size, random_normal_dimensions])
        generator_labels = tf.constant([[1.]] * batch_size)
        discriminator.trainable = False
        aloss, aacc = adversarial.train_on_batch(noise, generator_labels)
    
    score["d"].append([dloss, dacc])
    score["a"].append([aloss, aacc])
    display_html("<h4>{}</h>".format(
                 f"Epoch: {epoch + 1} "
                 f"A: loss,acc {score['a'][-1]} "
                 f"D: loss,acc {score['d'][-1]}"), raw=True)       
```

## Score:

```python
plt.figure(figsize=(10, 6))
plt.plot(np.array(score['d'])[:,:-1].flatten())
plt.plot(np.array(score['d'])[:,-1:].flatten())
plt.plot(np.array(score['a'])[:,:-1].flatten())
plt.plot(np.array(score['a'])[:,-1:].flatten())
plt.title("Losses & Accuracies")
plt.legend(["d loss", "d acc", "a loss", "a acc"])
plt.ylabel("Values")
plt.xlabel("Epochs")
plt.show()
```
